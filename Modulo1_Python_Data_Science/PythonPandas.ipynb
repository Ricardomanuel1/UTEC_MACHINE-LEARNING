```python
# Determinar la temperatura media de las ciudades de la UE y la temperatura
# media de las ciudades de fuera de la UE, antes y después del "Brexit".

# Es decir, para un par de promedios use los países actuales de la UE y los
# países no pertenecientes a la UE, y para el otro par pretenda que el
# Reino Unido no está en la UE.

#¡Imprima los cuatro números y asegúrese de etiquetar cuál es cuál!

#---

# Sugerencia: puede resolver esto en solo cinco líneas creando un "dataframe"
# unido y luego promediando las temperaturas para cuatro conjuntos diferentes
# de condiciones en el "dataframe".

# Nota: Para el 'or' de dos condiciones en pandas, use '|' en vez de '&'


citiesext = cities.merge(countries, on='country')

print('EU before Brexit:',\
      np.average(citiesext[citiesext.EU == 'yes'].temperature))
print('Non-EU before Brexit:',\
      np.average(citiesext[citiesext.EU == 'no'].temperature))
print('EU after Brexit:',\
      np.average(citiesext[(citiesext.EU == 'yes') &\
                           (citiesext.country != 'United Kingdom')].temperature))
print('Non-EU after Brexit:',\
      np.average(citiesext[(citiesext.EU == 'no') |\
                         (citiesext.country == 'United Kingdom')].temperature))
```

```python
# ¿Cuántas mujeres casadas mayores de 50 años se embarcaron en Cherbourg?
# Nota: 'first' es una función en Pandas, por lo que 'titanic.first' generará un error.
# use 'titanic['first'] en su lugar

result = titanic[titanic["first"].str.contains('Mrs')]\
          [(titanic.age > 50) & (titanic.embarked == "Cherbourg")]\
          .sort_values("last")
result
```

```python
from pandas._libs.algos import ensure_int64
# ¿A cuántos pasajeros les falta su edad y cuál es la tarifa
# promedio que pagan estos pasajeros?
# NOTA: Los valores que faltan en Pandas son nulos,
# se imprimen como 'NaN' y la función pd.isnull(v)
# verifica si un valor es nulo.

print("pasajeros sin edad:",titanic["age"].isnull().values.sum())
print('Promedio:',np.average(titanic[titanic.age.isnull()].fare))
```

```python
# ¿Cuál es la tarifa promedio pagada por los pasajeros en las tres clases,
# y la edad promedio de los pasajeros en dichas clases?

titanic.groupby(['class']).mean()[['fare','age']]

```

```python
# ¿Cuál fue la tarifa más alta pagada por un sobreviviente
# masculino, una sobreviviente femenina, un no sobreviviente
# masculino y una no sobreviviente femenina? (cuatro números)

titanic.groupby(['gender',"survived"]).max()[['fare']]
```

```python
from numpy.ma.core import count
# ¿Cuál es la tasa de supervivencia de los pasajeros en las tres clases,
# es decir, qué fracción de pasajeros en cada clase sobrevivió?
# ¿Cuál es la tasa de supervivencia de las mujeres frente a los hombres?
# ¿De niños (menores de 18 años) versus adultos (de 18 años o más)?

# NOTA: Las edades que faltan fallan automáticamente en las comparaciones
# como en SQL, y 'clase' es una función en Pandas, por lo que 'titanic.class'
# generará un error; use 'titanic['class'] en su lugar.

#print(titanic["survived"].value_counts())

print('First class:', len(titanic[(titanic['class'] == 1) & (titanic.survived == 'yes')])\
      /len(titanic[(titanic['class'] == 1)]))

print('Second class:', len(titanic[(titanic['class'] == 2) & (titanic.survived == 'yes')])\
      /len(titanic[(titanic['class'] == 2)]))

print('Third class:', len(titanic[(titanic['class'] == 3) & (titanic.survived == 'yes')])\
      /len(titanic[(titanic['class'] == 3)]))

print('mujeres/hombres:', len(titanic[(titanic['gender'] == "F") & (titanic.survived == 'yes')])\
      /len(titanic[(titanic['gender'] == "M") & (titanic.survived == 'yes')]))

print('niños/adultos:', len(titanic[(titanic['age'] < 18 ) & (titanic.survived == 'yes')])\
      /len(titanic[(titanic['age'] > 18) & (titanic.survived == 'yes')]))
```
